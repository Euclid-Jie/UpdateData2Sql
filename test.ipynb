# AKshare获取指数数据

```python
import akshare as ak

stock_zh_a_daily_qfq_df = ak.stock_zh_a_daily(
    symbol="sh000016", start_date="20250101", end_date="20250801"
)
data = stock_zh_a_daily_qfq_df[
    ["date", "open", "high", "low", "close", "volume", "amount"]
].rename(
    columns={
        "open": "OPEN",
        "high": "HIGH",
        "low": "LOW",
        "close": "CLOSE",
        "volume": "VLOUME",
        "amount": "AMT",
    }
)
data["PCT_CHG"] = data["CLOSE"].pct_change() * 100
```

# wind指数网站获取指数数据

```python
import requests
import pandas as pd

# https://www.windindices.com/indices/zh/IndexF9/e2d5a98547c3ee7c923a0259cee963e4
url = "https://indexapi.wind.com.cn/indicesWebsite/api/Kline?indexId=6644c422b6edae80b3c7a7d55803bc9e&period=1Y&lan=cn"
res = requests.get(url)
data = pd.DataFrame(res.json()["Result"]["data"])
data = data[["tradeDate", "open", "hight", "low", "close", "pctChange", "volume", "amount"]]
```

```python
data.rename(
    columns={
        "tradeDate": "date",
        "open": "OPEN",
        "hight": "HIGH",
        "low": "LOW",
        "close": "CLOSE",
        "pctChange": "PCT_CHG",
        "volume": "VOLUME",
        "amount": "AMT",
    }
)
```
